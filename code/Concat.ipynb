## pd.concat()
**pandas.concat(objs, *, axis=0, join='outer', ignore_index=False, keys=None, levels=None, 
names=None, verify_integrity=False, sort=False, copy=None)**

Useful when you want to concatenate the dataframes vertically or horizontally.

By default, pd.concat() concatenates the dataframes vertically. And you need to have datasets with the same column names
e.g. Monthly / Yearly datasets with the same formats

This is what will happen if you don't â†“

```python
pd.concat([steve,annie,marijke])
```

We need the same columns names to concat dataframes.

```python
# rename the columns so we have the same column names
marijke.columns = cols
steve.columns = cols
```

```python
pd.concat([steve,annie,marijke])
```

As you can see above, the index of the joined dataframe is also concatenated.
To reset the index, we can pass the **ignore_index = True** argument.

```python
joined = pd.concat([steve,annie,marijke], ignore_index = True)
```

```python
joined
```

```python
joined.dropna(how = 'any') 
# by default, how = "any", meaning rows that have any NULL value will be dropped.
# You can also specify how = "all", which will only remove rows that have NULL values for all the columns
```

```python
# reset the index (inplace: drop = True)
final = joined.dropna().reset_index(drop = True)
```

```python
final
```

### pd.concat(axis = 1)
To concatenate the dataframes horizontally, pass the **axis = 1** argument.

```python
pd.concat([marijke, annie, steve], axis = 1)
```

```python
marijke = marijke.dropna()
steve = steve.dropna()
```

```python
pd.concat([marijke, annie, steve], axis = 1)
```

```python
marijke = marijke.reset_index(drop = True)
steve = steve.reset_index(drop = True)
```

```python
pd.concat([marijke, annie, steve], axis = 1)
```

For example, you have the height and weight of every student and somehow you decide to calculate BMI using an API. If you get the result dataframe in the same order, you can just concat them horizontally.
Useful when you know the two dataframes are **in the same order** and **have the same index**, otherwise you need pd.merge() or pd.join().
